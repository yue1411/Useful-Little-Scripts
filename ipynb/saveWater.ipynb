```python
import numpy as np
import pandas as pd
import os
import glob
import fileinput
import re
pd.options.display.max_columns = 999
pd.options.display.max_rows = 999
pd.options.display.max_colwidth = 999
!pwd
```

```python
pdb_folder = '/Users/YueZhang/Documents/1_projects/8_cage/water_clustering/oh1_3water/'
clutering_folder = '/Users/YueZhang/Documents/1_projects/8_cage/water_clustering/'
```

```python
pdbs = glob.glob('%sRes*.pdb' %pdb_folder)

my_cols = ["ATOM", "ATOM_NUM", "ATOM_NAME", "RESIDUE_NAME", "CHAIN_NAME","IDENTIFIER","x", "y", "z", "1" , "0", "ATOM_TYPE"]

counter = 0

with open ('%sall_pdbs' %clutering_folder, "w") as outfile:
    for f in pdbs:  
        
        # replace A1 with A 1, A2 with A 2 to *.tmp, prevent mess up columns later 
        fin = open (f,"rt")
        fout = open ("%s%d.tmp" %(pdb_folder,counter),"wt")
        for line in fin:
            fout.write(line.replace('A1', 'A 1').replace('A2', 'A 2'))
        fin.close()
        fout.close()
        
        # save *.tmp as *.csv file, add which_pdb column
        df = pd.read_csv("%s%d.tmp" %(pdb_folder,counter), delim_whitespace=True, error_bad_lines=False,names = my_cols)
        df["which_pdb"]=counter
        export_csv = df.to_csv (r'%s%d.csv' %(pdb_folder,counter), index = None, header=True)
        
        # save this csv to all_pdb
        with open("%s%d.csv" %(pdb_folder,counter), "r") as infile:
            outfile.write(infile.read())
        
        
        #remove tmps and csvs
        tmp_name = '%d.tmp' %counter
        csv_name = '%d.csv' %counter
        tmp_path = os.path.join(pdb_folder, tmp_name)
        csv_path = os.path.join(pdb_folder, csv_name)
        os.remove(tmp_path)
        os.remove(csv_path)
        
        counter = counter + 1 

outfile.close()        
        
        
print "Save all pdbs to 'all_pdbs'."
print 'Total poses' , counter

```
